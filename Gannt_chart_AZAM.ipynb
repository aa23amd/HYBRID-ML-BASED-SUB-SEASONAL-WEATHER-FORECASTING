```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define project timeline (12 weeks starting from January 22, 2025)
start_date = datetime(2025, 1, 22)
tasks = [
    ("Literature Review & Dataset Exploration", start_date, start_date + timedelta(weeks=2), "Completed"),
    ("Baseline Model Implementation (XGBoost, LSTM)", start_date + timedelta(weeks=2), start_date + timedelta(weeks=4), "Completed"),
    ("Model Evaluation & Feature Engineering", start_date + timedelta(weeks=4), start_date + timedelta(weeks=6), "Ongoing"),
    ("Advanced Model Development (Transformers, Hybrid Models)", start_date + timedelta(weeks=6), start_date + timedelta(weeks=8), "Upcoming"),
    ("Model Fine-tuning & Comparative Analysis", start_date + timedelta(weeks=8), start_date + timedelta(weeks=10), "Upcoming"),
    ("Final Results, Interpretations, and Refinements", start_date + timedelta(weeks=10), start_date + timedelta(weeks=11), "Upcoming"),
    ("Report Writing & Final Submission", start_date + timedelta(weeks=11), start_date + timedelta(weeks=12), "Upcoming"),
]

# Convert tasks to a DataFrame
df = pd.DataFrame(tasks, columns=["Task", "Start", "End", "Status"])

# Plot Gantt Chart
fig, ax = plt.subplots(figsize=(10, 6))
colors = {"Completed": "green", "Ongoing": "orange", "Upcoming": "blue"}

for i, task in enumerate(df.itertuples()):
    ax.barh(task.Task, (task.End - task.Start).days, left=mdates.date2num(task.Start), color=colors[task.Status])

# Format chart
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %d"))
plt.xticks(rotation=45)
plt.xlabel("Timeline")
plt.ylabel("Tasks")
plt.title("Project Gantt Chart")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()

# Show the Gantt Chart
plt.show()
```
